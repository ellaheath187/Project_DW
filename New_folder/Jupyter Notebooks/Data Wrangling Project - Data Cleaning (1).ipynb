```python
import pandas as pd 
```

# Data Cleaning

```python
seed_records_df = pd.read_csv('seed_records.csv')
seed_records_df['SEED'] = seed_records_df['SEED'].astype(str).str.replace("No. ", "")

# Split Overall Record Column
seed_records_df[['WINS', 'LOSSES']] = seed_records_df['OVERALL RECORD'].str.split('-', expand=True).astype(int)
#Split Best Finish Column
seed_records_df[['BEST FINISH TYPE', 'BEST FINISH COUNT']] = seed_records_df['BEST FINISH'].str.split('(', n=1, expand=True)
seed_records_df['BEST FINISH TYPE'] = seed_records_df['BEST FINISH TYPE'].str.strip()
seed_records_df['BEST FINISH COUNT'] = seed_records_df['BEST FINISH COUNT'].str.replace('times', '')\
                                                          .str.replace(')', '')\
                                                          .str.strip()
#Make Best Finish Count numeric 
seed_records_df['BEST FINISH TYPE'] = seed_records_df['BEST FINISH TYPE'].replace('Champions', 'Champion')
seed_records_df['BEST FINISH COUNT'] = seed_records_df['BEST FINISH COUNT'].replace('six', 6)
seed_records_df['BEST FINISH COUNT'] = seed_records_df['BEST FINISH COUNT'].replace('five', 5)
seed_records_df['BEST FINISH COUNT'] = seed_records_df['BEST FINISH COUNT'].replace('four', 4)
seed_records_df['BEST FINISH COUNT'] = seed_records_df['BEST FINISH COUNT'].replace('two', 2)
seed_records_df['BEST FINISH COUNT'] = seed_records_df['BEST FINISH COUNT'].replace('once', 1)
# Convert Seed and Best Finish Count to integers
seed_records_df['SEED'] = seed_records_df['SEED'].astype(int)
seed_records_df['BEST FINISH COUNT'] = seed_records_df['BEST FINISH COUNT'].astype(int)

display(seed_records_df[['SEED', 'OVERALL RECORD', 'WINS', 'LOSSES','BEST FINISH' , 'BEST FINISH TYPE' , 'BEST FINISH COUNT']])

seed_records_df.to_csv('seed_records.csv', index=False)
```

```python
seed_pct_df = pd.read_csv('seed_pct_records.csv')
seed_pct_df['SEED VS. SEED.'] = seed_pct_df['SEED VS. SEED.'].str.replace("*", "")
seed_pct_df[['Seed_1', 'Seed_2']] = seed_pct_df['SEED VS. SEED.'].str.split('vs.', expand=True)
seed_pct_df['Seed_1'] = seed_pct_df['Seed_1'].astype(int)
seed_pct_df['Seed_2'] = seed_pct_df['Seed_2'].astype(int)
seed_pct_df[['Wins', 'Losses']] = seed_pct_df['W-L'].str.split('-', expand=True)
seed_pct_df['Wins'] = seed_pct_df['Wins'].astype(int)
seed_pct_df['Losses'] = seed_pct_df['Losses'].astype(int)
display(seed_pct_df[['SEED VS. SEED.' , 'Seed_1' , 'Seed_2', 'W-L' ,'Wins', 'Losses', 'PCT.' ]])

seed_pct_df.to_csv('seed_pct_records.csv', index=False)
```
